```python
import pandas as pd # This line imports the pandas library, essential for data manipulation and analysis
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Load your dataset (replace 'your_dataset.csv' with the actual dataset file path)
df = pd.read_csv("C:/Users/MSI/OneDrive - ESPRIT/Bureau/AB_NYC_2019.csv")
df
```

```python
# 1. Identify variable types
print("\n--- Variable Types ---")
print(df.dtypes)

# Separate qualitative (categorical) and quantitative (numerical) variables
categorical_vars = df.select_dtypes(include=['object', 'category']).columns
numerical_vars = df.select_dtypes(include=['number']).columns

print("\nCategorical Variables:", categorical_vars.tolist())
print("Numerical Variables:", numerical_vars.tolist())
```

```python
    # 2. Detect missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())

# Visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# 3. Identify outliers and understand distributions
print("\n--- Outlier Detection and Distributions ---")
for col in numerical_vars:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df[col].dropna(), kde=True, bins=30, color='blue')
    plt.title(f"Histogram of {col}")
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color='orange')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# 4. Check correlations between variables
print("\n--- Correlation Matrix ---")
correlation_matrix = df[numerical_vars].corr()
print(correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Additional insights: Pairwise relationships
sns.pairplot(df[numerical_vars].dropna(), diag_kind="kde", corner=True)
plt.show()

print("\n--- Analysis Completed ---")
```

```python
# Step 2: Data Preparation
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,,LabelEncoder, OrdinalEncoder

from sklearn.utils import resample
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# 2.1 Handle missing data
print("\n--- Handling Missing Data ---")
# Impute missing values for numerical data using mean
numerical_vars = df.select_dtypes(include=['number']).columns
imputer_mean = SimpleImputer(strategy='mean')
df[numerical_vars] = imputer_mean.fit_transform(df[numerical_vars])

# Impute missing values for categorical data using mode
categorical_vars = df.select_dtypes(include=['object', 'category']).columns
imputer_mode = SimpleImputer(strategy='most_frequent')
df[categorical_vars] = imputer_mode.fit_transform(df[categorical_vars])
print("Missing values handled successfully.")
```

```python
#2.2 
#Identification of Qualitative Columns
categorical_vars = df.select_dtypes(include=['object', 'category']).columns
print("\nQualitative (Categorical) Columns:", categorical_vars.tolist())
ordinal_cols = []  
label_cols = []    
onehot_cols = list(set(categorical_vars) - set(ordinal_cols) - set(label_cols))
```

```python
# Encoding with OneHotEncoder
if onehot_cols:
print("\n--- Encodage avec OneHotEncoder ---")
onehot_encoder = OneHotEncoder(sparse=True, drop='first')  # drop='first' pour Ã©viter la multicolinÃ©aritÃ©
onehot_encoded = onehot_encoder.fit_transform(df[onehot_cols])  # Appliquer OneHotEncoder
#Convert Sparse Matrix to DataFrame and Add to Original DataFrame
onehot_df = pd.DataFrame.sparse.from_spmatrix(onehot_encoded, columns=onehot_encoder.get_feature_names_out(onehot_cols))
df = pd.concat([df, onehot_df], axis=1)
df.drop(columns=onehot_cols, inplace=True)
```

```python
# Encoding with LabelEncoder
if label_cols:
print("\n--- Encodage avec LabelEncoder ---")
label_encoder = LabelEncoder()
for col in label_cols:
df[col] = label_encoder.fit_transform(df[col])
```

```python
# Encoding with OrdinalEncoder 
if ordinal_cols:
print("\n--- Encodage avec OrdinalEncoder ---")
ordinal_mapping = { "column_name": ['Category1', 'Category2', 'Category3']  }
ordinal_encoder = OrdinalEncoder(categories=[ordinal_mapping[col] for col in ordinal_cols])
df[ordinal_cols] = ordinal_encoder.fit_transform(df[ordinal_cols])
```

```python
# Displaying data after encoding
print("\n--- DonnÃ©es aprÃ¨s encodage ---")
print(df.head())
# Save the encoded file 
df.to_csv("AB_NYC_2019_encoded.csv", index=False)
print("\nFichier encodÃ© sauvegardÃ© : AB_NYC_2019_encoded.csv")
```
